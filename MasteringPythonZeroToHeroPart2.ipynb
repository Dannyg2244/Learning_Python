```python
"""Let say you are a teacher and you have different students 
records containing the id of a student and the marks list of each student
where different students have taken different numbers of subjects.
All these records are in hard copy. You want to enter all the data in computer
and want to compute the average marks of each students and display"""

def studentRecord():
    D = {}
    while True:
        studentId = input("enter student id: ")
        markList = input("enter the students score separated by coma: ")
        moreStudent = input('more students? type "no" to exit')
        if (studentId in D):
            print("Error, this student Id", studentId, "existed already")
        else:
            D[studentId] = markList#.split(",")
        if (moreStudent.lower() == "no"):
            return D
```

```python
studentRecord()
```

```python
def takeAvgMark(D):
    avgMarks = {}
    for x in D:
        L = D[x]
        S = 0
        for marks in L:
            S += int(marks) 
            avgMarks[x] = S/len(L)
    return avgMarks
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('C:/covid/covid_19_data.csv')
```

```python
df.head(2)
```

```python
#df.drop(['SNo', 'Last Update'], axis=1, inplace=True)
df.rename(columns={'ObservationDate':'Date', 'Province/State':'State', 'Country/Region':'Country'}, inplace=True)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head(2)
```

```python
imputer = SimpleImputer(strategy='constant')
df2 = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
```

```python
df3 = df2.groupby(['Country', 'Date'])[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

```python
df3.head(20)
```

```python
countries = df['Country'].unique()
```

```python
countries
```

```python
for idx in range(0, len(countries)):
    C = df3[df3['Country']==countries[idx]].reset_index()
    plt.scatter(np.arange(0, len(C)), C['Confirmed'], color='blue', label='Confirmed')
    plt.scatter(np.arange(0, len(C)), C['Deaths'], color='green', label='Deaths')
    plt.scatter(np.arange(0, len(C)), C['Recovered'], color='red', label='Recovered')
    plt.title(countries[idx])
    plt.xlabel('Days since the first suspect')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
```

```python
df4 = df3.groupby(['Date'])[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

```python
C = df4
plt.scatter(np.arange(0, len(C)), C['Confirmed'], color='blue', label='confirmed')
plt.scatter(np.arange(0, len(C)), C['Deaths'], color='green', label='Deaths')
plt.scatter(np.arange(0, len(C)), C['Recovered'], color='red', label='Recovered')
plt.title('World')
plt.xlabel('Days since the first suspect')
plt.ylabel('Number of cases')
plt.legend()
plt.show()
```
